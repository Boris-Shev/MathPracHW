```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
years = data[:,0]
years = years.astype('int32')
hare = data[:,1]
lynx = data[:,2]
carr = data[:,3]
```

# 1)  Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
fig = plt.figure()
plt.figure(figsize=(10, 5))
fig.patch.set_facecolor('xkcd:white')
X = years

plt.plot(X, hare, label='Hares')
plt.plot(X, lynx, label='Lynxes')
plt.plot(X, carr, label='Carrots')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Population')
plt.xticks(X[::2],fontsize=10, rotation=20)
plt.legend()
plt.grid()
plt.show()
```

# 2) Find which years any of the populations is above 50000. Show time periods where populations are above 50000 in the plot of populations.

```python
fig = plt.figure()
plt.figure(figsize=(10, 5))
fig.patch.set_facecolor('xkcd:white')
X = years

plt.plot(X, hare, label='Hares')
plt.plot(X, lynx, label='Lynxes')
plt.plot(X, carr, label='Carrots')
plt.plot([1900, 1920],[50000, 50000], label='50k bound')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Population')
plt.xticks(X[::2],fontsize=10, rotation=20)
plt.legend()
plt.grid()
plt.scatter(X[np.where(hare>5e4)[0]], hare[hare>5e4], s=50)
plt.scatter(X[np.where(lynx>5e4)[0]], lynx[lynx>5e4], s=50)
plt.scatter(X[np.where(carr>5e4)[0]], carr[carr>5e4], s=50)
plt.show()
```

# 3) Find which year each species had the largest population. Mark these years in the plot of populations.

```python
fig = plt.figure()
plt.figure(figsize=(10, 5))
fig.patch.set_facecolor('xkcd:white')
X = years
Spec = [hare, lynx, carr]

plt.plot(X, hare, label='Hares')
plt.plot(X, lynx, label='Lynxes')
plt.plot(X, carr, label='Carrots')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Population')
plt.xticks(X[::2],fontsize=10, rotation=20)
plt.legend()
plt.grid()
plt.scatter(X[np.argmax(Spec,1)], np.max(Spec,1), s=50, c='r')
plt.show()
```

# 4) Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
fig = plt.figure()
plt.figure(figsize=(10, 5))
fig.patch.set_facecolor('xkcd:white')
X = years
sum = hare + lynx

plt.bar(X, 100 * hare / sum,  bottom= 100 * lynx / sum, label='hare')
plt.bar(X, 100 * lynx / sum, label='lynx')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Population')
plt.xticks(X[::2],fontsize=10, rotation=20)
plt.legend()
plt.grid()
plt.show()
```

# 5) Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
plt.figure(figsize=(15, 6))
fig.patch.set_facecolor('xkcd:white')
X = years

plt.subplot(131)
plt.hist(hare, color='b')
plt.plot([np.mean(hare), np.mean(hare)] ,[0, 6],c='r')
plt.title("mean = " + str(np.mean(hare)) + "\nstd = " + str(np.std(hare)))

plt.subplot(132)
plt.hist(lynx, color='g')
plt.plot([np.mean(lynx), np.mean(lynx)] ,[0, 6],c='r')
plt.title("mean = " + str(np.mean(lynx)) + "\nstd = " + str(np.std(lynx)))

plt.subplot(133)
plt.hist(carr, color='orange')
plt.plot([np.mean(carr), np.mean(carr)] ,[0, 6],c='r')
plt.title("mean = " + str(np.mean(carr)) + "\nstd = " + str(np.std(carr)))

plt.show()
```

# 6) Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

```python
plt.figure(figsize=(15, 6))
fig.patch.set_facecolor('xkcd:white')
X = years

a = np.empty(21, dtype=str)
a[np.where(hare / lynx > 1)[0]] = 'H'
a[np.where(hare / lynx <= 1)[0]] = 'L'
print(a)
plt.pie([np.count_nonzero(a == 'H'), np.count_nonzero(a == 'L')], labels=['H','L'],
       shadow=True)

plt.show()
```

# 7) Plot the change in the hare population and the number of lynxes. Find the correlation coefficient.

```python
plt.figure(figsize=(15, 6))
fig.patch.set_facecolor('xkcd:white')
X = years
plt.bar(X[1:], hare[1:] - hare[:-1], width=-0.4, align='edge')
plt.bar(X[1:], lynx[1:] - lynx[:-1],  width=0.4, align='edge')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Changes', fontsize=10)
plt.xticks(X[::1],fontsize=10, rotation=20)
plt.grid()

plt.show()
```

# 8) Using a scatter plot, show the population of hares vs carrots and hares vs lynxes by year on two separate subplots (each point corresponds to a particular year, and the point coordinates correspond to the population of the two species in that year).

```python
plt.figure(figsize=(15, 6))
fig.patch.set_facecolor('xkcd:white')
X = years

plt.subplot(121)
plt.scatter(hare, carr, c='orange')
plt.title('Hares vs Carrots',fontsize=16)
plt.xlabel('Hares', fontsize=14)
plt.ylabel('Carrots',fontsize=14)

plt.subplot(122)
plt.scatter(hare, lynx)
plt.title('Hares vs Lynxes',fontsize=16)
plt.xlabel('Hares', fontsize=14)
plt.ylabel('Lynxes', fontsize=14)

plt.show()
```

# 9) Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value with the true value and print a ratio of the error to the true value.

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(years[:-1].reshape(-1,1), hare[:-1])
res = regressor.predict(years[-1].reshape(-1,1))[0]
print("True value =", hare[-1])
print("Estimated value =", res)
print("Ratio of the error to the true value =", np.abs(res - hare[-1]) / hare[-1])


```
